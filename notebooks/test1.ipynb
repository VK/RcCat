```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
filename = r"D:\LOG00030.TXT"
```

```python
lineData = []

with open(filename, 'r') as fp:
    line = fp.readline()
    cnt = 1
    while line:
        values = line.split("\t")
        if(len(values) == 12):
            try:
                ldata = np.array(values).astype(np.int32)
                lineData.append(ldata)
            except:
                pass

        if(line.startswith("#F") and len(values) == 4):
            print(line)
        #print("Line {}: {}".format(cnt, line.strip()))
        line = fp.readline()
        cnt += 1

lineData = np.array(lineData)
```

```python
plt.figure(figsize=(15,5))

"""
plt.plot(lineData[:, 2], label="x")
plt.plot(lineData[:, 3], label="y")
plt.plot(lineData[:, 4], label="z")
"""
plt.plot(lineData[:, 5], label="a_tot")


plt.plot(lineData[:, 6]/100, label="pitch")
plt.plot(lineData[:, 7]/100, label="roll")
plt.plot(lineData[:, 8], label="dist")

#plt.plot(np.sqrt(( lineData[:, 2]**2 + lineData[:, 3]**2 + lineData[:, 4]**2)))

plt.plot(lineData[:, 11]*100, label="state")

plt.xlim([9000,17000])

plt.legend()
```
